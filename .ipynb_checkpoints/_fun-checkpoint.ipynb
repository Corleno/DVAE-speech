```python
import IPython.display as ipd
import os
import numpy as np
import matplotlib.pyplot as plt
import torch
import soundfile as sf
import librosa
```

```python
# STFT
wlen_sec=64e-3
hop_percent=0.25
fs=16000
zp_percent=0
trim=True
verbose=False

# number of fft is supposed to be a power of 2
nfft = np.int(np.power(2, np.ceil(np.log2(wlen_sec * fs))))
# window length is equal to nfft by default 
wlen = nfft
# hop: number of audio samples between adjacent STFT columns
hop = np.int(hop_percent * wlen)
# a vector or array of length nfft (sin function, 0 ~ pi)
win = np.sin(np.arange(0.5, wlen+0.5) / wlen * np.pi)
```

```python
file = ('/Users/xiaoyu/WorkStation/Project_rvae/Data/clean_speech/wsj0_si_et_05/444/444c0211.wav')
y, sr = librosa.load(file)
ipd.Audio(data=y, rate=sr)
```

```python
x, fs_x = sf.read(file)
x = x / np.max(np.abs(x))
X = librosa.stft(x, n_fft=nfft, hop_length=hop,win_length=wlen, window=win) 
data_orig = np.abs(X) ** 2
print(data_orig.shape)
```
